```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline

dataset = pd.read_csv('time_series_2019-ncov-Confirmed (1).csv')
data = dataset.drop(['Province/State','Lat','Long'],axis=1)
```

```python
c = data.keys()
c = c[1:]
```

```python
for i in c:
    a = data[i]
    j = 0
    count = 0
    while j < len(a):
        if a[j] == 0:
            count+=1
        j+=1
    if count >= 340:
        data = data.drop([i],axis=1)
key = data.keys()
```

```python
df1 = data.iloc[:,0]
df1 = np.array(df1)
df = data.values
```

```python
for ii in range(0,487):
    aa = df1[ii]
    for jj in range(ii,487):
        bb = df1[jj] 
        if aa == bb:
            if ii == jj:
                continue
            else:
                for qq in range(1,20):
                    df[ii,qq] = df[ii,qq] + df[jj,qq]
                    df[jj,qq] = 0
                
        else:
            continue
```

```python
d = pd.DataFrame(df,columns=key)
```

```python
m = d.iloc[:,1:]
m = m.mean(axis=1)
```

```python
for i in range(0,487):
    if m[i] == 0:
        d = d.drop([i],axis=0)
d
```
