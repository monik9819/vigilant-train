```python
#!/usr/bin/env python3

import re
import operator
import csv

def rankingError():
    error_dict = dict()
    final_error = list()
    with open("syslog.log", "rt") as fh:
        for line in fh:
            line=line.strip()
            res = re.search(r"ticky: ERROR ([\w ']*) ", line)
            if res == None:
                continue
            error_dict[res.groups()[0]] = error_dict.get(res.groups()[0],0) + 1
    error_lst = sorted(error_dict.items(), key = operator.itemgetter(1), reverse=True)
    for i,j in error_lst:
        final_error.append({"Error": i,"Count":j})
    with open("error_message.csv","w") as fh_csv:
        writer = csv.DictWriter(fh_csv, fieldnames=["Error","Count"])
        writer.writeheader()
        writer.writerows(final_error)

def userUsageStats():
    usage_info = dict()
    usage_error = dict()
    usernames = list()
    final_usage = list()
    with open("syslog.log", "rt") as fh:
        
        for line in fh:
            line=line.strip()
            if re.search(r"INFO [\w ]* \[#\d{1,5}\] \(([\w].*)\)$", line):
                res = re.search(r"INFO [\w ]* \[#\d{1,5}\] \(([\w.]*)\)$", line)
                usage_info[res.groups()[0]] = usage_info.get(res.groups()[0],0) + 1
                
            elif re.search(r"ERROR [\w ']* \(([\w.]*)\)$", line):
                res = re.search(r"ERROR [\w ']* \(([\w.]*)\)$", line)
                usage_error[res.groups()[0]] = usage_error.get(res.groups()[0],0) + 1
            
            res = re.search(r"\(([\w.]*)\)$", line)
            usernames.append(res.groups()[0])
            
    usernames = sorted(list(set(usernames)))
    for username in usernames:
        try:
            xf = usage_info[username]
        except KeyError:
            xf = 0
        try:
            xe = usage_error[username]
        except KeyError:
            xe = 0
        final_usage.append({"Username":username, "INFO":xf, "ERROR":xe})
    with open("user_statistics.csv","w") as fh_csv:
        writer = csv.DictWriter(fh_csv, fieldnames=["Username", "INFO", "ERROR"])
        writer.writeheader()
        writer.writerows(final_usage)

rankingError()
userUsageStats()
```
