### mapping 
use <code>map</code> and <code>lambda</code> functions

```python
applicant_income_mean = final["ApplicantIncome"].mean() #found mean applicants income
xfinal = final["ApplicantIncome"].map(lambda p: p- applicant_income_mean) #mapped it with every customer 
xfinal #saved in series format
```

###### if want to add the mapped values directly to the dataframe use <code>apply</code> method

```python
def remean_points(row):
    row.ApplicantIncome = row.ApplicantIncome - applicant_income_mean #change done is particular cell
    return row #then after operation return the value of entire row to form dataframe

final.apply(remean_points, axis='columns').head()
```

###### Ques.  There are only so many words you can use when describing a bottle of wine. Is a wine more likely to be "tropical" or "fruity"? Create           a Series descriptor_counts counting how many times each of these two words appears in the description column in the dataset.








```python
n_trop = reviews.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])
descriptor_counts
```

###### Ques. We'd like to host these wine reviews on our website, but a rating system ranging from 80 to 100 points is too hard to understand - we'd like to translate them into simple star ratings. A score of 95 or higher counts as 3 stars, a score of at least 85 but less than 95 is 2 stars. Any other score is 1 star.
<h6>Also, the Canadian Vintners Association bought a lot of ads on the site, so any wines from Canada should automatically get 3 stars, regardless of points.</h6>
<h6>Create a series star_ratings with the number of stars corresponding to each review in the dataset</h6>

```python
def myfunc(row):
    if row.country == "Canada":
        row.points = 3
        return row
    elif row.points >= 95:
        row.points = 3
        return row
    elif (row.points < 95) and (row.points > 84):
        row.points = 2
        return row
    else:
        row.points = 1
        return row

reviews.apply(myfunc,axis="columns")
```
