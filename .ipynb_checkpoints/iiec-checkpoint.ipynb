###### for python recording it

###### for recognition of voice
<code>speech_recognition.Recognizer()</code>

```python
import speech_recognition as sr
```

```python
#to recongize voice means to convert frequency into words
#recognizer has many different recognizer API as methods on it we can even use to do the task mentioned above
#
r = sr.Recognizer()
```

```python
#speech_recognition.Microphone() internally uses pyaudio to listen through mic

with sr.Microphone() as source: #using with allows us to stop recording as we stop automatically
    print("starting talking")
    audio = r.listen(source)
    print("speech done...")
```

```python
#data saved in audio is in binary form...the frequency are saved in binary in computer
print(type(audio),"\n")
audio.frame_data
```

```python
r.recognize_google(audio) #runs the binary data with google api to convert it into words from 0's and 1's
```
