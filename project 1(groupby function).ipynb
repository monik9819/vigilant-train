## data cleaning and visulization


```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline

dataset = pd.read_csv('time_series_2019-ncov-Confirmed (1).csv')
data = dataset.drop(['Province/State','Lat','Long'],axis=1)
```

```python
c = data.keys()
c = c[1:]
```

```python
for i in c:
    a = data[i]
    j = 0
    count = 0
    while j < len(a):
        if a[j] == 0:
            count+=1
        j+=1
    if count >= 340:
        data = data.drop([i],axis=1)
```

```python
col1 = data.iloc[:,0].unique()
col1.sort()
```

```python
d = data.groupby(['Country/Region']).sum()
print(d.shape)
d.head()
```

```python
plt.figure(figsize=(100,50))
ax=plt.subplot(1,1,1)
x=range(171)
y=d.iloc[:,-1]
ax.set_xticks(range(171))
ax.set_xticklabels(col1, rotation=82,fontsize=18)
plt.yticks(fontsize=40)
plt.plot(x,y,lw=5)
plt.title('Country wise plot',fontsize=70)
plt.xlabel('Countries',fontsize=70)
plt.ylabel('Total COVID Cases',fontsize=60)
plt.show()
```

```python
col = d.columns
summa=[]
for i2 in col:
    summa.append(d[i2].sum())
```

```python
plt.figure(figsize=(60,30))
ax=plt.subplot(1,1,1)
x=range(len(col))
y=summa
plt.yticks(fontsize=35)
ax.set_xticks(x)
ax.set_xticklabels(col, rotation=36, fontsize=35)
plt.plot(x,y, marker='o',c='blue', markersize=30)
plt.title('Datewise wise cases',fontsize=70)
plt.xlabel('Dates',fontsize=70)
plt.ylabel('Total COVID Cases',fontsize=55)
plt.show()
```
