# make turtle animation 

```python
import turtle

ww =turtle.Screen()      #create a graphic window
ww.bgcolor("lightgreen") #set the window background color

ales = turtle.Turtle()   #create named ales
ales.color("blue")       #make tess blue
ales.pensize(10)          #set the width of her pen

ales.forward(100)        #moves forward by 100 units (100 pixels)
ales.left(90)            #turns left by 90 degrees
ales.forward(150) 

ww.exitonclick()         #exis the window as user click on canvas
```

```python
#make a square spiral

import turtle

win = turtle.Screen()
win.bgcolor("black")

slo = turtle.Turtle()
slo.color("white")
slo.pensize(7)

dist = 30

for i in range(10):
    for j in range(3):
        slo.forward(dist)
        slo.left(90)
    dist+=10
    
win.exitonclick()
```

```python

import turtle

wn = turtle.Screen()
wn.bgcolor("lightgreen")

tess = turtle.Turtle()
tess.color("blue")
tess.shape("turtle")   #can define shape of pointer

dist = 5
tess.up()              #lifts the pen up without making any impression
for _ in range(30):    
    tess.stamp()       #leave an impression on the canvas
    tess.forward(dist) 
    tess.right(24)     
    dist = dist + 2
    
wn.exitonclick()
```

```python
#makes a ring of turtles face to each other
import turtle

wn = turtle.Screen()
wn.bgcolor("lightgreen")

tess = turtle.Turtle()
tess.color("blue")
tess.shape("turtle")

dist = 30
tess.up()                     # this is new
for _ in range(14):    # start with size = 5 and grow by 2
    tess.stamp()                # leave an impression on the canvas
    tess.forward(dist)          # move tess along
    tess.right(24)              # and turn her


wn.exitonclick()
```

```python
#makes a ring face outwards
import turtle

wn = turtle.Screen()
wn.bgcolor("black")

alex = turtle.Turtle()
alex.color("aliceblue")
alex.shape("turtle")
alex.speed(1)
alex.up()

for size in range(10):
    alex.forward(50)
    alex.stamp()
    alex.forward(-50)
    alex.right(36)
    
alex.forward(300)
    
wn.exitonclick()
```

```python
#makes circular spiral
import turtle
wn = turtle.Screen()
wn.bgcolor("lightgreen")
tess = turtle.Turtle()
tess.color("blue")
tess.shape("turtle")

dist = 20
tess.up()                     # this is new
for _ in range(30):    # start with size = 5 and grow by 2
    tess.stamp()                # leave an impression on the canvas
    tess.forward(dist)          # move tess along
    tess.right(24)              # and turn her
    dist = dist + 2
wn.exitonclick()
```
